```python
def moving_window(tweet):
    L1=[]                                       #initializing an empty list to store three-phrase words
    L2=[]                                       #initializing an empty list to store value 1
    for index in range(len(tweet)):             #creating a string of three words
        if index<=len(tweet)-3:
            string1=' '.join(tweet[index:index+3])
            L1.append(string1)                  #adding the three word strings to list L1
    for idx in range(len(L1)):
        L2.append(1)                            #adding 1 (value) to list L2
    tuple_phrases=tuple(zip(L1,L2))             #creating a tuple from Lists L1 and L2 (to convert to dictionary)
    phrases=dict((x,y) for x,y in tuple_phrases) #creating a dictionary 
    return phrases
```

Part (b)

```python

def cosine(tweet1,tweet2):
    matches=0                                   #initializing matches to 0
    n1=len(moving_window(tweet1))               #initializing n1 to no of three word phrases in 1st tweet
    n2=len(moving_window(tweet2))               #initializing n2 to no of three word phrases in 1st tweet
    for key in moving_window(tweet1):
        if key in moving_window(tweet2):        
            matches=matches+1                   #incrementing matches if a same phrase is found in 2nd tweet
    x=matches/math.sqrt(n1*n2)                  #computing cosine function
    return x
```

```python
def word_count(n):  # Function to count the number times a word is used in a line.
    counts = dict()
    words=convertWords(n) # Convert input line into the list of words using previously written function.

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts  #Return a dictionary having keys as words and values as the number of times specific word is used.

def postings_lists(b):
    newDic=dict()
    newDic2=word_count(b) #Store dictionary of word counts.
    newDic[str(b)]=newDic2 # Create a dictionary having key a quote and value a dictionary of word counts.
    
    return newDic
FinalDic=dict()
for b in range(0,len(newList),1):
    newDic=dict()
    newDic=postings_lists(newList[b])
    FinalDic=FinalDic | newDic      #Create a postings list dictionary for all quotes.

print(FinalDic)

def posting2(newList):  # Another posting list function with different input type. To be used in the coming sections.
    posting = {}
    for j in newList:
        words = convertWords(j)
        quote = {}
        for w in words:
            if w in quote:
                quote[w] += 1
            else: 
                quote[w] = 1
        posting[j] = quote
    return posting
```
